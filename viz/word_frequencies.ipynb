```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
word_with_freqs = pd.read_csv("../data/word_frequency.csv", header=None, names=["word", "freq"])
```

```python
def plot_word_frequencies(word_freqs):
    plt.figure(figsize=(10, 6))
    plt.bar(word_freqs["word"], word_freqs["freq"])
    plt.xlabel("Word")
    plt.ylabel("Frequency")
    plt.title("Word Frequencies")
    plt.xticks(rotation=60)
    plt.tight_layout()
    plt.show()
```

```python
sorted_word_freqs = word_with_freqs.sort_values(by="freq", ascending=False)
```

```python
top_20_words = sorted_word_freqs.head(20)
```

```python
plot_word_frequencies(top_20_words)
```

```python
print(sorted_word_freqs.reset_index().head(4000).to_string(float_format="{:.16f}".format))
```

Around top 3000 words, commonality of words are stagnant (frequency is in single digits and not changing often) and then there is a long tail of words that are very uncommon. This is consistent with Zipf's law, which states that in natural language, the frequency of any word is inversely proportional to its rank in the frequency table.
